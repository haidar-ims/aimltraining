```python
df=pd.read_excel("students.xlsx")
```

```python
df
```

```python
top_students=df[df["Marks"]>80]
top_students
```

```python
rejected_students=df[df["Marks"]<80]
rejected_students
```

This line saves your top_students table into an Excel file called selected_students.xlsx in the current folder.
*to_excel() is a pandas function that saves your DataFrame(df) into an Excel file.
*"selected_students.xlsx" is the filename — it creates a file with that name.
*index=False means don’t include the row numbers (0, 1, 2, 3...) in the Excel file.
You usually don’t need them when saving nice, clean data.

```python
top_students.to_excel("selected_students.xlsx", index=False)
print("Filtered data saved to 'selected_students.xlsx'")
```

```python
rejected_students.to_excel('rejected_students.xlsx',index=False)
print('File Created')
```

```python
data=pd.read_csv("Final_data.csv")
```

```python
data
```

# Filter data of persons whose weight is greater than 60 Kg. We want only Age, Gender, Weight, protein_per_kg

```python
selected_data=data[data['Weight (kg)']>60] 
final_data=selected_data[['Age','Gender','Weight (kg)','protein_per_kg']]
final_data
```

## Optional One Line Coding ##
final_data = data[data['Weight (kg)'] > 60][['Age', 'Gender', 'Weight (kg)', 'protein_per_kg']]
final_data

```python
final_data.to_csv("overweight.csv", index=False)
print("Filtered data saved to 'overweight.csv'")
```

# Use patients file and display all records
## Show number of rows and columns in your dataset

```python
patients=pd.read_csv("patients.csv")
```

```python
print('Patients Record as Follows')
patients
```

# Create the list of patients who are in the ICU
## save the file with the name icu_patients.csv

```python
icu_patients = patients[patients['service'] =='ICU'][['name', 'service']]
icu_patients
```

```python
icu_patients.to_csv("icu_patients.csv", index=False)
print("Filtered data saved to 'icu_patients.csv'")
```
